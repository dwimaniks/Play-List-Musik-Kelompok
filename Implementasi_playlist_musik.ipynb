Doubly linked list

```python
class DNode:
    """Node untuk doubly linked list"""
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None  # berguna untuk operasi di akhir (akses cepat)

    def append(self, data):
        """Tambahkan node di akhir"""
        node = DNode(data)
        if not self.head:           # list kosong
            self.head = self.tail = node
        else:
            self.tail.next = node
            node.prev = self.tail
            self.tail = node

    def insert_after(self, key, data):
        """
        Sisipkan node baru setelah node yang memiliki nilai == key.
        Jika key tidak ditemukan, tidak melakukan apa-apa (bisa disesuaikan).
        """
        cur = self.head
        while cur:
            if cur.data == key:
                new_node = DNode(data)
                nxt = cur.next
                # hubungkan new_node dengan cur dan nxt
                cur.next = new_node
                new_node.prev = cur
                new_node.next = nxt
                if nxt:
                    nxt.prev = new_node
                else:
                    # jika cur adalah tail, update tail
                    self.tail = new_node
                return True
            cur = cur.next
        return False

    def delete_by_key(self, key):
        """Hapus node pertama yang datanya == key"""
        cur = self.head
        while cur:
            if cur.data == key:
                # jika node adalah head
                if cur.prev:
                    cur.prev.next = cur.next
                else:
                    self.head = cur.next
                # jika node adalah tail
                if cur.next:
                    cur.next.prev = cur.prev
                else:
                    self.tail = cur.prev
                # bersihkan pointer pada cur (optional)
                cur.prev = cur.next = None
                return True
            cur = cur.next
        return False

    def to_list(self):
        """Kembalikan daftar data dari head ke tail (untuk menampilkan)"""
        res = []
        cur = self.head
        while cur:
            res.append(cur.data)
            cur = cur.next
        return res

dll = DoublyLinkedList()
# buat 3 node awal: A, B, C
dll.append("A")
dll.append("B")
dll.append("C")
print("Awal (3 node):", dll.to_list())   # ['A', 'B', 'C']

# sisipkan 'X' setelah 'A' (penyisipan tengah)
dll.insert_after("A", "X")
print("Setelah insert X setelah A:", dll.to_list())  # ['A','X','B','C']

# hapus node dengan key 'B'
dll.delete_by_key("B")
print("Setelah delete key 'B':", dll.to_list())     # ['A','X','C']

# hapus node head 'A'
dll.delete_by_key("A")
print("Setelah delete key 'A':", dll.to_list())     # ['X','C']

# hapus node tail 'C'
dll.delete_by_key("C")
print("Setelah delete key 'C':", dll.to_list())     # ['X']
```

Circular linked list

```python
class CNode:
    """Node untuk circular linked list (single link cukup)"""
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append_end(self, data):
        """Sisipkan node di akhir (sebagai tail), mempertahankan sifat circular"""
        new_node = CNode(data)
        if not self.head:
            # satu node, next menunjuk ke dirinya sendiri
            self.head = new_node
            new_node.next = new_node
        else:
            # temukan tail (node yang next-nya menunjuk ke head)
            tail = self.head
            while tail.next is not self.head:
                tail = tail.next
            tail.next = new_node
            new_node.next = self.head

    def delete_last(self):
        """Hapus node terakhir (tail). Jika kosong -> False. Jika satu node-> jadi kosong."""
        if not self.head:
            return False
        # jika hanya satu node
        if self.head.next == self.head:
            self.head = None
            return True
        # cari node sebelum tail (node whose next.next == head)
        prev = None
        cur = self.head
        while cur.next is not self.head:
            prev = cur
            cur = cur.next
        # sekarang cur adalah tail, prev adalah sebelum tail
        prev.next = self.head
        return True

    def to_list(self):
        """Kembalikan list dari circular list (mulai dari head). Jika kosong -> []"""
        res = []
        if not self.head:
            return res
        cur = self.head
        while True:
            res.append(cur.data)
            cur = cur.next
            if cur == self.head:
                break
        return res

cll = CircularLinkedList()
# buat 3 node awal: 1, 2, 3
cll.append_end(1)
cll.append_end(2)
cll.append_end(3)
print("Awal circular (3 node):", cll.to_list())   # [1,2,3]

# sisipkan di akhir: 4
cll.append_end(4)
print("Setelah append 4 di akhir:", cll.to_list()) # [1,2,3,4]

# hapus node terakhir
cll.delete_last()
print("Setelah delete_last:", cll.to_list())      # [1,2,3]

# hapus sampai kosong
cll.delete_last()  # now [1,2]
cll.delete_last()  # now [1]
cll.delete_last()  # becomes []
print("Setelah hapus sampai kosong:", cll.to_list()) # []
```
